```python
import numpy as np
# convert the response variables to numeric 0/1 from binary N/Y

y = y.apply(lambda x: np.where(x == 'N',0,1))
y
```

```python
# splitting the dataset into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
# calculate the naive accuracy, precision and recall
# from sklearn.model_selection import cross_val_score

# naive accuracy
accuracy = np.sum(y) / len(y)
print(accuracy)
```

```python
# implementing Logistic Regression with GridSearchCV to identify the best fitting model
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import scorer, accuracy_score, fbeta_score, make_scorer
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    # ('vect', TfidfVectorizer(stop_words='english')),
    ('clf', LogisticRegression())
])

parameters = {
    # 'vect__max_df': (0.25, 0.5, 0.75),
    # 'vect__stop_words': ('english', None),
    # 'vect__max_features': (2500, 5000, 10000, None),
    # 'vect__ngram_range': ((1, 1), (1, 2)),
    # 'vect__use_idf': (True, False),
    # 'vect__norm': ('l1', 'l2'),
    'clf__penalty': ('l1', 'l2'),
    'clf__C': (0.01, 0.1, 1, 10),
}
```

```python
grid_search = GridSearchCV(pipeline, parameters, n_jobs=-1, verbose=1, scoring='accuracy', cv=3)
```

```python
grid_search.fit(X_train, y_train)
```

```python
# clf = LogisticRegression()
# parameters = {'C': [10**-i for i in range(-5, 5)], 'class_weight': [None, 'balanced']}
# scorer = make_scorer(fbeta_score,beta=.5)
# grid_obj = GridSearchCV(estimator=clf,param_grid=parameters,scoring=scorer)
```

```python
# grid_fit = grid_obj.fit(X_train,y_train)
grid_fit = grid_search
```

```python
# get the best estimator
best_clf = grid_fit.best_estimator_
y_best_pred = best_clf.predict(X_test)
```

```python
# predictions from the unoptimized model
clf = LogisticRegression()

y_pred = (clf.fit(X_train, y_train)).predict(X_test)
```

```python
# Comparing the accuracy and fscore of the two models

print('Unoptimized Model:')
print('---------------------')
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('F-Score: ', fbeta_score(y_test, y_pred, beta=0.5))
print('---------------------')
print('Optimized Model:')
print('---------------------')
print('Accuracy: ', accuracy_score(y_test, y_best_pred))
print('F-Score: ', fbeta_score(y_test, y_best_pred, beta=0.5))
print('---------------------')
```
